```r
library(ggplot2)
```

```r
files = list.files()
```

```r
files = files[grepl('csv',files)]
```

```r
data = list()
for(f in files){
    data[[f]]=read.csv(f)
    data[[f]]$file = f
    data[[f]]$epoch = 1:nrow(data[[f]])
}

df = do.call(rbind,data)
```

```r
options(repr.plot.width = 10, repr.plot.height = 4)

m=reshape2::melt(df,id.vars=c('epoch','file'))

colors = c(AA.train='firebrick1',
           AA.val='darkred',
           AB.train = 'deepskyblue',
           AB.val='royalblue',
           AC.train='orange',
           AC.val='darkorange4')

m$chrom = unlist(strsplit(m$file,'\\.'))[seq(1,nrow(m)*5,5)]
m$seq_len = gsub('seqLen\\.','',unlist(strsplit(m$file,'\\.'))[seq(2,nrow(m)*5,5)])
m$group = gsub('group','',unlist(strsplit(m$file,'\\.'))[seq(3,nrow(m)*5,5)])

m$dataset = 'train'
m[grepl('val',m$variable),'dataset']='val'
m$variable = gsub('val_','',m$variable)

m$dataset2 = paste(m$group,m$dataset,sep='.')
m$chrom_len = paste(m$chrom,m$seq_len,sep='.')

```

```r
ggplot(subset(m,seq_len=='seqLen800'),aes(epoch,as.numeric(value),col=dataset2))+
    geom_line(aes(linetype=factor(dataset)))+facet_wrap(variable~.,ncol=4,scales='free')+
    theme_bw()+
    scale_color_manual(values=colors)
```
